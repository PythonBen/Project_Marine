```python
def func_path_folder(N=0):
    """ return the full path to a given directory"""

    path_data = Path('/media/ben/Ubuntu 20_04_4 LTS amd64/submeeting_2022/submeeting_images/total_images_su/')
   
    if N==0:
        folder = 'boulouris2_frames_0_3525_320x240_SU'
    elif N==1:
        folder = 'pyramide2_frames_0_900_step_25_resized_320x240'
    elif N==2:
        folder = 'pyramide3_frames_0_380_320x240_SU'
    elif N==3:
        folder = 'pyramide4_frames_0_900_320x240_SU'

    path_folder = path_data/folder

    return path_folder
```

```python
def files_in_folders(N_folders=4):
    """ return a dictionary with the folder name and the number of files in it """
    
    outputdict={}
    for N in range(N_folders):
        count=0
        path_folder = func_path_folder(N)
        for f in path_folder.iterdir():
            count+=1
        mydict = {path_folder.name:count}
        outputdict.update(mydict)
    return outputdict

mydict = files_in_folders()
print(mydict)
```

```python
def selecting_images(tmin, tmax, path_input, labels, thres=0.8):
    """ fonction to select images in a given time interval
    this is used to order images by classes
    thres = 0.8 means 20 % for validation set and 80 % for train set"""
    pattern = r"(\d+)"
    path_name = Path('/submeeting_images/')
    new_path = path_name/path_input.name
    list_select = []
    list_all = [f.stem for f in path_input.iterdir()]
    for e in list_all:
        res = re.match(pattern, e)
        if res is not None:
            time = int(res.group(0))
      
        if (time >= tmin) and (time <= tmax):
            list_select.append(e)
    is_valid = []
    for i in range(len(list_select)):
        random_number = random.uniform(0, 1)
        if random_number <= thres:
            is_valid.append(False)
        else:
            is_valid.append(True)

    dict1 = {'fname': list_select, 'labels': labels, 'is_valid': is_valid}
    df = pd.DataFrame(dict1)
    return df
```

```python
# one hot encoding
def one_hot_label_encoder(dataframe):
   
    str_to_int = {"algen":0,"divers":1,"robots":2,"rocks":3,"sand":4,"structure":5,"water_only":6}
    Y_onehot = []
    #Y_onehot = torch.zeros((len(dataframe),3))
    for i in range(len(dataframe)):
        labels = dataframe.iloc[i]['labels'].split(' ')

        labels_coded = torch.LongTensor([str_to_int[label] for label in labels])
    
        y_onehot = nn.functional.one_hot(labels_coded, num_classes=7)
        
        y_onehot = y_onehot.sum(dim=0).float()
        y_onehot = y_onehot.numpy()
        Y_onehot.append(y_onehot)
   
        
    return Y_onehot
```

```python
df_encoded['labels_encoded'] = one_hot_label_encoder(df_encoded)
```
